# ASSIMENT 1

```python
import numpy as np
```

```python
# Create a 1D array with integers from 1 to 20
arr_1d = np.arange(1, 21)

# a. Sum, mean, median, standard deviation
print("Sum:", np.sum(arr_1d))
print("Mean:", np.mean(arr_1d))
print("Median:", np.median(arr_1d))
print("Standard Deviation:", np.std(arr_1d))

# b. Indices of elements greater than 10
indices = np.where(arr_1d > 10)
print("Indices of elements > 10:", indices)
```

```python
# Create a 2D array of shape 4x4 with values 1 to 16
arr_2d = np.arange(1, 17).reshape(4, 4)

# a. Print the array
print("Original Array:\n", arr_2d)

# b. Transpose of the array
print("Transpose:\n", arr_2d.T)

# c. Row-wise and Column-wise sums
print("Row-wise Sum:", np.sum(arr_2d, axis=1))
print("Column-wise Sum:", np.sum(arr_2d, axis=0))
```

```python
# Create two 3x3 arrays with random integers between 1 and 20
a = np.random.randint(1, 21, size=(3, 3))
b = np.random.randint(1, 21, size=(3, 3))

print("Array A:\n", a)
print("Array B:\n", b)

# a. Element-wise operations
print("Addition:\n", a + b)
print("Subtraction:\n", a - b)
print("Multiplication:\n", a * b)

# b. Dot product
print("Dot Product:\n", np.dot(a, b))
```

```python
# Create a 1D array and reshape to 3x4
arr = np.arange(1, 13).reshape(3, 4)

print("Reshaped Array (3x4):\n", arr)

# Slice first two rows and last two columns
sliced = arr[:2, -2:]
print("Sliced [First 2 Rows, Last 2 Columns]:\n", sliced)
```

# ASSIMENT 2

```python
import pandas as pd
```

```python
# Create DataFrame from given data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
df
```

```python
print(df.head())
```

```python
# b. Summary statistics
print(df[['Age', 'Salary']].describe())
```

```python
# c. Average salary in HR department
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average HR Salary:", hr_avg_salary)
```

```python
df['Bonus'] = df['Salary'] * 0.10
df
```

```python
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_by_dept)
```

```python
sorted_df = df.sort_values(by='Salary')
sorted_df.to_csv('sorted_employees.csv', index=False)
sorted_df
```

```python
# !pip install matplotlib  

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='blue')

plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure(figsize=(8, 5))
plt.bar(students, marks, color=['skyblue', 'lightgreen', 'salmon', 'orange'])

plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Highlight the region with the highest revenue
explode = [0.1 if val == max(revenue) else 0 for val in revenue]

plt.figure(figsize=(7, 7))
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.title("Company Revenue Distribution by Region")
plt.show()
```

```python
# Generate 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color='purple', edgecolor='black')

plt.title("Histogram of Random Integers (1 to 100)")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
